```python
# download_images.py
import os, requests, time
import pandas as pd
from tqdm import tqdm

def download_image(url, path, retries=3):
    for _ in range(retries):
        try:
            r = requests.get(url, timeout=10)
            if r.status_code == 200:
                with open(path, 'wb') as f:
                    f.write(r.content)
                return True
        except:
            time.sleep(1)
    return False

def download_all(csv_path, out_dir):
    df = pd.read_csv(csv_path)
    os.makedirs(out_dir, exist_ok=True)
    for _, row in tqdm(df.iterrows(), total=len(df)):
        img_url = row['image_link']
        img_path = os.path.join(out_dir, f"{row['sample_id']}.jpg")
        if not os.path.exists(img_path):
            success = download_image(img_url, img_path)
            if not success:
                print("Failed:", img_url)

download_all('dataset/train.csv', 'images/train/')
download_all('dataset/train.csv', 'images/test/')
https://m.media-amazon.com/images/I/51oPFLIsvDL.jpg
```
